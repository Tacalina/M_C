```python
#Funci칩n para encontrar que intervalos tienen el 6/4 (Garantizando que sea 4 justa)


def encontrar_64(dataframe):

  """
  Funci칩n que me devuelve los indices de las filas que contienen los acordes
  6/4, donde se verifica que la cuarta sea justa(P4) y la sexta sea mayor o menor (M6 o m6)

  Parameters
  ----------
  dataframe : DataFrame
      DataFrame con todos los acordes de la partitura

  Returns
  -------
  indices : list
      Lista con los indices de las filas que cumplen con la condici칩n de tener
      el acorde 6/4.
  
  df_filtrado : DataFrame
      DataFrame con los acordes que cumplen con la condici칩n de tener el acorde 6/4
  """
  
  nuevalista = dataframe['Nombre intervalo simplificado']
  indices = []
  for indice, lista in enumerate(nuevalista):
    if ('M6' in lista or 'm6' in lista) and 'P4' in lista:
      
      indice_fila = indice
      indices.append(indice_fila)

    df_filtrado = df.iloc[indices]

  return indices, df_filtrado
```

```python
#Funci칩n encontrar voces dentro del 6/4 

def voces_64(df):

    """
    Funci칩n que encuentra las voces que contienen el intervalo 6/4 en un dataframe de pandas

    Parameters
    ----------
    df : dataframe
        Dataframe de pandas con informaci칩n de los acordes de la partitura

    Returns

    n_indice4 : list
        Lista con los 칤ndices de las voces que contienen el intervalo de cuarta justa
    n_indice6 : list
        Lista con los 칤ndices de las voces que contienen el intervalo de sexta mayor o menor

        
    """
    nuevalista = df['Nombre intervalo simplificado']
    n_indice4 = []
    n_indice6 = []
    a = df['Nombre intervalo simplificado'].tolist()
    for j in a:
        
            voz = j
            indice4 = voz.index('P4') +1
            
            if 'm6' in voz:
                indice6 = voz.index('m6') + 1
            if 'M6' in voz:
                indice6 = voz.index('M6') + 1
            n_indice4.append(indice4)
            n_indice6.append(indice6)
        

    return n_indice4 , n_indice6
```

##### Funci칩n para verificar las condiciones de 6/4 de paso 

```python
import numpy as np
from music21 import *

def condiciones_paso(bajos, BB_n, BA_n, N4sb, N6sb, V4b, V6b, V4a, V6a):
    """
    Esta funci칩n calcula los intervalos entre bb y nb y ba y nb, y verifica dos condiciones.

    Par치metros:
    bajos, BB_n, BA_n, N4sb, N6sb, V4b, V6b, V4a, V6a: listas de notas o valores relacionados con las notas.

    Devuelve:
    seg_m_b, seg_m_a, cond3, cond4: numpy arrays con los resultados de las verificaciones.

    suma: numpy array con la suma de las verificaciones.
    
    """

    inter_BA = []
    inter_BB = []
    cond3 = []
    cond4 = []

    for i in range(len(bajos)):

        bb = BB_n[i][-1]   # Nota del bajo antes del acorde 6/4, con BB != NB 
        ba = BA_n[i][-1]    # Nota del bajo despu칠s del acorde 6/4, con BA != NB
        nb = bajos[i]       # Nota del bajo del acorde 6/4

        n4sb = N4sb[i]   # Nota de la voz de la cuarta del acorde 6/4 
        n6sb = N6sb[i]   # Nota de la voz de la sexta del acorde 6/4

        n4bb = V4b[i][-1]  # Nota de la cuarta antes del acorde 6/4
        n6bb = V6b[i][-1]  # Nota de la sexta antes del acorde 6/4

        n4ba = V4a[i][-1]  # Nota de la cuarta despu칠s del acorde 6/4
        n6ba = V6a[i][-1]  # Nota de la sexta despu칠s del acorde 6/4

        #Definir las notas con music21

        bb = note.Note(bb)  
        ba = note.Note(ba)
        nb = note.Note(nb)

        #Definir los intervalos con music21

        int_b = interval.notesToInterval(bb,nb)
        int_a = interval.notesToInterval(ba,nb)

        #Agregar los intervalos a las listas 
        inter_BA.append(int_a.semiSimpleName)
        inter_BB.append(int_b.semiSimpleName)

        if (bb<nb and ba>nb) or (bb>nb and ba<nb):   #Condici칩n 3
            cond3.append(True)
        else:
            cond3.append(False)

        if (n4sb == n4bb) and (n4sb == n4ba):   #Condici칩n 4
            cond4.append(True)
        else:
            cond4.append(False)

    seg_m_b = [i == 'm2' for i in inter_BB]  #Verificar si son segunda menor (primera condici칩n)
    seg_m_a = [i == 'm2' for i in inter_BA]  #Verificar si son segunda menor (segunda condici칩n)

    #Convertir a numpy array
    seg_m_b, seg_m_a = np.array(seg_m_b), np.array(seg_m_a) 
    cond3 = np.array(cond3)   
    cond4 = np.array(cond4)

    #Sumar todas las condiciones

    suma = np.sum([cond3,cond4,seg_m_a,seg_m_b],axis=0)

    return seg_m_b, seg_m_a, cond3, cond4, suma
```
